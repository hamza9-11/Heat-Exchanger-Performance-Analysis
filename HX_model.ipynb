```python
import pandas as pd
df = pd.read_csv('heat_exchanger_dataset.csv', sep=',')
df.head()
```

```python
df.describe()
```

```python
original_cols = [col for col in df.columns if 'noisy' not in col]
correlations = df[original_cols].corr()['hx_1_heat_load_kw'].sort_values(ascending=False)
print(correlations)
```

```python
df_clean = df[['hot_inlet_temperature_k', 'cold_inlet_mass_flow_kg_s', 'hx_1_heat_load_kw']].copy()
df_clean.columns = ['temp', 'flow', 'target']

X_clean = df_clean[['temp', 'flow']]
y_clean = df_clean['target']

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
df_noisy = df[['hot_inlet_temperature_k_noisy', 'cold_inlet_mass_flow_kg_s_noisy', 'hx_1_heat_load_kw_noisy']].copy()
df_noisy.columns = ['temp', 'flow', 'target']

Q1 = df_noisy.quantile(0.25)
Q3 = df_noisy.quantile(0.75)
IQR = Q3 - Q1
mask = ~((df_noisy < (Q1 - 3 * IQR)) | (df_noisy > (Q3 + 3 * IQR))).any(axis=1)
df_noisy_final = df_noisy[mask]

X_noisy_test = df_noisy_final[['temp', 'flow']]
y_noisy_test = df_noisy_final['target']

X_noisy_test_scaled = scaler.transform(X_noisy_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)

```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_val_pred = model.predict(X_val_scaled)

mae_clean = mean_absolute_error(y_val, y_val_pred)
r2_clean = r2_score(y_val, y_val_pred)

print(f" Performance on Clean Data (Baseline): ")
print(f"Mean Absolute Error (MAE): {mae_clean:.4f} kW")
print(f"R-squared Score (R2): {r2_clean:.4f}")
```

```python
y_noisy_pred = model.predict(X_noisy_test_scaled)

mae_noisy = mean_absolute_error(y_noisy_test, y_noisy_pred)
r2_noisy = r2_score(y_noisy_test, y_noisy_pred)

print(f"Performance on Noisy Data:")
print(f"MAE: {mae_noisy:.4f} kW")
print(f"R2 Score: {r2_noisy:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

sns.scatterplot(x=y_val, y=y_val_pred, ax=ax1, alpha=0.6)
ax1.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--', lw=2, label='Ideal')
ax1.set_title(f'Performance on Clean Data\n(R2: {r2_clean:.4f})', fontsize=12)
ax1.set_xlabel('Actual Heat Load (kW)')
ax1.set_ylabel('Predicted Heat Load (kW)')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.6)

sns.scatterplot(x=y_noisy_test, y=y_noisy_pred, ax=ax2, alpha=0.6, color='orange')
ax2.plot([y_noisy_test.min(), y_noisy_test.max()], [y_noisy_test.min(), y_noisy_test.max()], 'r--', lw=2, label='Ideal')
ax2.set_title(f'Performance on Noisy Data\n(R2: {r2_noisy:.4f})', fontsize=12)
ax2.set_xlabel('Actual Heat Load (kW)')
ax2.set_ylabel('Predicted Heat Load (kW)')
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

rf_model.fit(X_train_scaled, y_train)

y_noisy_rf_pred = rf_model.predict(X_noisy_test_scaled)

mae_rf = mean_absolute_error(y_noisy_test, y_noisy_rf_pred)
r2_rf = r2_score(y_noisy_test, y_noisy_rf_pred)

print(f"--- Random Forest Performance ---")
print(f"MAE: {mae_rf:.4f} kW")
print(f"R2 Score: {r2_rf:.4f}")

improvement_rf = r2_rf - r2_noisy
print(f"Improvement over Linear Model: +{improvement_rf:.4f}")
```

```python
plt.figure(figsize=(12, 6))

subset = 100

plt.plot(y_noisy_test.values[:subset], label='Actual Data (Noisy)', color='black', alpha=0.5, linestyle='--')

plt.plot(y_noisy_pred[:subset], label=f'Linear Model (R2={r2_noisy:.2f})', color='red', linewidth=1.5)

plt.plot(y_noisy_rf_pred[:subset], label=f'Random Forest (R2={r2_rf:.2f})', color='blue', linewidth=2)

plt.title('Linear vs Random Forest on Noisy Data', fontsize=14)
plt.xlabel('Sample Index')
plt.ylabel('Heat Load (kW)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
